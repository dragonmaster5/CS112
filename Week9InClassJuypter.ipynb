Problem 3. The array expenses contains the amount spent in a week:
expenses = np . array ([50 , 40 , 60 , 30 , 80 , 70 , 100])

(1) Calculate the total expenses for the week.

(2) Calculate the average daily expense.

(3) Create a new array that shows the difference between each day’s expense and the
average daily expense.

(4) Suppose that the person wants to save some money to buy a new computer and he
sets the daily limit to be 80. How many days did he need to spend more than the
limit?


```python
expenses=np.array([50 , 40 , 60 , 30 , 80 , 70 , 100])
totalExpenses=sum(expenses)
print(f"Total Expenses: {totalExpenses}")
avgE= totalExpenses/len(expenses)
print(f"Average spent per day: {avgE.round(2)}")
diffE= expenses-avgE
print(f"The difference from the average expenses is {diffE.round(2)}")
print(f" how many days over 80: {len(expenses[expenses>80])}")
```

Problem 5. Consider the following numpy array which records the temperature in various
days in Chicago in July 2024.

temps = np . array ([72 , 75 , 80 , 85 , 78 , 88 , 93 , 77 , 70 , 65])

(1) Create a new numpy array containing all temperatures that are above 80.

(2) Create a new numpy array containing all temperatures that are between 70 and 72
(inclusive).

(3) Create a new NumPy array containing temperatures that are either above 90 (kind
of warm) or below 68 (kind of cold).

(4) Find the mean, the standard deviation, and the range of this data set.

(5) Create a new numpy array containing all temperatures that are below the mean

```python
temps = np . array ([72 , 75 , 80 , 85 , 78 , 88 , 93 , 77 , 70 , 65])
tempsG80=temps[temps>80]
print(f"Temps Greater than 80: {tempsG80}")
tempsB=temps[(temps>=70) & (temps<=72)]
print(f"Temps between 70 and 72 (inclusive): {tempsB}")
tempKinda= temps[(temps>90) | (temps<68)]
print(f"Temps that are kinda warm or kinda cold: {tempKinda}")
print(f"Mean: {temps.mean()}")
print(f"Standard Deviations: {temps.std().round(2)}")
print(f"Range: {temps.max()-temps.min()}")
belowMeanTemp= temps[temps<temps.mean()]
print(f"Temps below the mean: {belowMeanTemp}")
```
